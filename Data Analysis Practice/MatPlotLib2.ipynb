```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y)
plt.xlabel("Madrchod")
plt.title("Not Cool")
plt.show()
```

```python
plt.xlabel("Madrchod")
```

```python
plt.subplot(1,2, 2)
plt.plot(x, y)
plt.subplot(1,2,1)
plt.plot(y, x)
```

```python
fig = plt.figure()
axes = fig.add_axes([1.0, 1.0, 1.0, 1.0])

axes.plot(x, y)
axes.set_xlabel("Gay")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.6])
axes2 = fig.add_axes([0.2, 0.3, 0.3, 0.3])

axes1.plot(x, y)
axes2.plot(y, x, 'r--')
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Cool Plot")
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
# axes[0].plot(x, y, 'r*')
# axes[1].plot(y, x, 'b--')
for ax in axes:
    ax.plot(x, y, 'g*')
    
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(20,4), dpi=100)
axes = fig.add_axes([0.1, 0.1, 0.4, 0.5])
axes.plot(x, y,'r--')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(20,20))
for ax in axes:
    for a in ax:
        a.plot(x, y, 'g--')
        a.set_xlabel("X Label")
        a.set_ylabel("Y Label")
```

```python
fig.savefig("filename.png")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.2, 0.9, 0.9])
axes.plot(x, y, 'r--', label="x, y")
axes.plot(y, y, 'g', label="y, x", marker=".")
axes.legend()
```
