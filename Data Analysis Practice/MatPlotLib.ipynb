```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

#### FUNCTIONAL METHOD OF CREATING A MATPLOTLIB PLOT

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 4)
plt.plot(y, x, 'b')
```

#### OBJECT ORIENTED

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left bottom width height from 0 to 1
axes2 = fig.add_axes([1, 0.1, 0.8, 0.8])
# axes3 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_title('LEFT PLOT')
axes2.plot(y, x)
axes2.set_title('RIGHT PLOT')
plt.show()
```

### Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(y, x, 'g')
fig.tight_layout()
```

```python
fig, axes
```

#### Figure Size, Aspect Ratio and DPI

```python
# fig = plt.figure(figsize=(8, 2), dpi=100) #figsize=tuple(width, height)
# ax = fig.add_axes([0, 0, 1, 1])
# ax.plot(x, y)
fig, axes = plt.subplots(nrows=2, ncols=1,  figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.show()
plt.tight_layout()
```

## Customizing the plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="gray", linewidth=3, alpha=0.5, linestyle='-', marker="o", markersize='10', 
       markerfacecolor="yellow", markeredgewidth="2", markeredgecolor="green") #color = name, hexcodes #linewidth = number of times of default #alpha = brightness #linestyle='--' 
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
