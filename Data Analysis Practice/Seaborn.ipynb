# Seaborn
A data vizualization library built on matplotlib

### Distribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

**DISPLOT**<br>
Used to show the distribution of a univariate(1 variable) set of observations
The variable chosen vs it's count

```python
sns.distplot(tips['total_bill'], kde=False, bins=30) 
```

**JOINTPLOT**<br>
Basically for combining 2 different distribution plots, it is bivariate in nature

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

**PAIRPLOT**<br>
Is used to plot pairwise relationships across an entire dataset, at least for the numerical columns, support color hue for categorical columns. <br>
Categorical columns are basically the ones that 2 or 3 categories in them, like sex has male and female 

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
```

**Rugplots**<br>
Used to explain KDE plots. This is basically the same as histograms but here, instead of a vertical stacking, a line is made for each point. So the more dense a section is, the more values are present there. <br><br>
KDE stands for Kernel Density Estimation plots 

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots
Used to plot categorical data 

### Bar Plots 
Most Basic categorical plots
Used to aggregate the categorical data based of some fucntion which by default is mean

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.mean)
```

### CountPlot is same as bar plot but it marks against the count

```python
sns.countplot(x='sex',data=tips)
```

### Box and Violin plots
Used to show distribution of categorical data

```python
fig = plt.figure(figsize=(10, 5))
axes = fig.add_axes([0, 0, 1, 1])
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
fig = plt.figure(figsize=(15, 5))
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### Strip Plot
Simplest, like scatter plot where 1 variable is categorical

```python
fig = plt.figure(figsize=(12, 5))
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### Swarm Plot 
Is a combination of both, violin and strip plots

It's similar to strip plot but the points are adjusted to prevent overlapping

Not recommended as most people don't know about it and don't it's slow for large datasets

```python
fig = plt.figure(figsize=(12, 5))
sns.swarmplot(x='day', y='total_bill', data=tips)
```

To combine violin and swarm plot, do the following way

again not recommended as it's too much information

```python
fig = plt.figure(figsize=(12, 5))
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

### Factor Plot (now called catplot)


```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
axes.xaxis.label.set_color('red')
```

## Matrix Plots
For heatmaps

```python
flights = sns.load_dataset('flights')
flights.head()
```

in order for a heat map to work properly, the data should be in a matrix form, this means the index name and the column name match up so that the cell value indicates something that is relevant to both of those names, something like the correlation between the 2 or something like that

```python
tc = tips.corr()
tc
```

```python
fig = plt.figure(figsize=(12, 8))
sns.heatmap(data=tc, annot=True, cmap="coolwarm")
```

```python
flights
```

```python
fl_pvt = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
tips.head()
```

```python
fig = plt.figure(figsize=(12,5))
sns.heatmap(data=fl_pvt, cmap="magma", linecolor="white", linewidth="3")
```

## Cluster Map
Uses hierarchical clustering to produce a clustered version of the heat map

```python
sns.clustermap(fl_pvt)
```

## Grids
To automate subplots based off features in the data

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris['species'].nunique()
```

```python
 sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
```

```python
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
# g.map(sns.histplot, 'total_bill')
```

```python
g.map(plt.scatter, 'total_bill', 'tip')
```
